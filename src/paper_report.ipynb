```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib_venn import venn2
```

```python
ven = pd.read_excel('/Users/navid/Documents/1_Projects/0_Age-It/Our Tasks/Mario_report/data/venezia_papers.xlsx')
keep = ['Contact Presenter Full Name', 'Title',
        'Keywords', 'Abstract']
ven = ven[keep]
ven.columns = ['authors', 'title', 'keywords', 'abstract']
```

```python
ven['title'] = ven['title'].apply(lambda x : x.lower().strip())
```

```python
ven.head()
```

```python
nap = pd.read_excel('/Users/navid/Documents/1_Projects/0_Age-It/Our Tasks/Mario_report/data/napoli_papers.xlsx')
keep = ['authors_formatted', 'title', 'abstract', 'keywords']
nap = nap[keep]
nap.columns = ['authors', 'title', 'abstract', 'keywords']
nap = nap[['authors', 'title', 'keywords', 'abstract']]
```

```python
nap['title'] = nap['title'].apply(lambda x : x.lower().strip())
```

```python
nap.head()
```

```python


# ven, nap must exist and both have a 'title' column
s1 = set(ven["title"].astype(str).str.strip().str.lower().dropna())
s2 = set(nap["title"].astype(str).str.strip().str.lower().dropna())

plt.figure(figsize=(6, 6))
venn2([s1, s2], set_labels=("Venezia", "Napoli"))
plt.title("Overlap of titles")
plt.show()

# (optional) print counts
print("ven unique titles:", len(s1))
print("nap unique titles:", len(s2))
print("intersection:", len(s1 & s2))
print("only ven:", len(s1 - s2))
print("only nap:", len(s2 - s1))
```
